```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as crl
```

```python
im = cv2.imread("cat.jpg")
plt.axis("off")
plt.imshow(im)
plt.show
```

   RGB:

```python
im_RGB = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(im_RGB)
plt.show()
```

   3  : RED, GREEN, BLUE

```python
R = im_RGB.copy()
```

```python
R[:,:,(1,2)] = 0
plt.axis("off")
plt.imshow(R)
plt.show
```

```python
G = im_RGB.copy()
```

```python
G[:,:,(0,2)] = 0
plt.axis("off")
plt.imshow(G)
plt.show
```

```python
B = im_RGB.copy()
```

```python
B[:,:,(0,1)] = 0
plt.axis("off")
plt.imshow(B)
plt.show
```

 ,    HSV :

```python
img_HSV = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
```

 :

```python
H =  img_HSV.copy()
H[:,:,(1,2)] = 255
H_RGB = cv2.cvtColor(H, cv2.COLOR_HSV2RGB)
plt.axis("off")
plt.imshow(H_RGB)
plt.show
```

```python
S =  img_HSV.copy()
S[:,:,(0,2)] = 180, 255
S_RGB = cv2.cvtColor(S, cv2.COLOR_HSV2RGB)
plt.axis("off")
plt.imshow(S_RGB)
plt.show
```

```python
V =  img_HSV.copy()
V[:,:,(0,1)] = 180, 0
V_RGB = cv2.cvtColor(V, cv2.COLOR_HSV2RGB)
plt.axis("off")
plt.imshow(V_RGB)
plt.show
```

    Lab:

```python
im_Lab = cv2.cvtColor(im, cv2.COLOR_BGR2LAB)
```

```python
L = im_Lab.copy()
L[:,:,(1,2)] = 80
L_RGB = cv2.cvtColor(L, cv2.COLOR_LAB2RGB)
L_Gray= cv2.cvtColor(L_RGB, cv2.COLOR_RGB2GRAY)
plt.axis("off")
plt.imshow(L_Gray, cmap='gray')
plt.show
```

```python
a = im_Lab.copy()
a[:,:,(0,2)] = 127, 0
a_RGB = cv2.cvtColor(a, cv2.COLOR_LAB2RGB)
a_GRAY = cv2.cvtColor(a_RGB, cv2.COLOR_RGB2GRAY)
```

```python
cmap_a = crl.LinearSegmentedColormap.from_list('custom blue', ['Red', 'Gray', 'Green'], N=255)
plt.axis("off")
plt.imshow(a_GRAY, cmap=cmap_a)
plt.show()
```

```python
b = im_Lab.copy()
b[:,:,(0,1)] = 127, 0
b_RGB = cv2.cvtColor(b, cv2.COLOR_LAB2RGB)
b_GRAY = cv2.cvtColor(b_RGB, cv2.COLOR_RGB2GRAY)
```

```python
cmap_b = crl.LinearSegmentedColormap.from_list('custom blue', ['Yellow', 'Gray', 'Blue'], N=255)
plt.axis("off")
plt.imshow(b_GRAY, cmap=cmap_b)
plt.show()
```
